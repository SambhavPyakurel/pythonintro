#### Q1. What is Statistics?

Statistics is the science of collecting, organizing and analyzing data.

#### Q2. Define the different types of statistics and give an example of when each type might be used.

1. Descriptive Statistics: It deals with organization and summarization a data.
2. Inferential Statistics: It deals with the analysis of data and drawing conclusion.

*e.g.*
Suppose height of students of a school (10 from each class) is taken.

Calculating mean, mode, median, Variance etc. of the heights and illustrating them in charts falls under Descriptive Statistics.

On the other hand, Using the sample heights and projecting the average height of students of each class falls under inferential statistics.

#### Q3. What are the different types of data and how do they differ from each other? Provide an example of each type of data.

1. Quantitative data: Numerical data such as height, weight of a person, population of city etc.
2. Qualitative data: Categorical data such as Gender, Race, User rating of an app etc.

Quantitative data is further divided into:
1. Discrete data: Data that can be counted. e.g. Total students in a school, Number of voters etc.
2. Continuous data: Data that cannot be counted but measured. Height, Weight etc.

Qualitative data is further divided into:
1. Nominal data: Categorical data where order or ranking is not meaningful. e.g. Gender, Race etc.
2. Ordinal data: Categorical data where order or ranking makes sense. e.g. User rating,School Grades etc.

#### Q5. Explain the concept of levels of measurement and give an example of a variable for each level.

Levels of measurement of data are tool used to categorize and describe data based on their characteristics and properties.

1. Nominal: Categorical data that have no order falls under this measure. e.g. Gender, Race
2. Ordinal: Categorical data that have order but the difference between two data can't be measured falls under this measure. e.g. Ratings, Education level.
3. Interval: Those Dataset whose difference can be measured but base lowest point is not zero(i.e. true zero doesn't exist). e.g Temperature in Celcius,Fahrenheit.
4. Ratio: Dataset where difference is measurable and lowest point is zero as well(i.e. true zero exists). e.g. Test Score, Age.

#### Q7. How nominal data type is different from ordinal data type.

Nominal data are the categorical data where order or ranking doesn't matter. e.g. Gender,Race.

Whereas, in Ordinal data the order or ranking of categorical data matters. e.g. User rating, Grades(classes).

In this way nominal data is different from ordinal data.

#### Q10. What are some common measures of central tendency and variability used in statistics? Explain how each measure can be used to describe a dataset.

The common measure of central tendency and variability used in statistics are:
1. Mean: It gives the average value of the dataset. Suppose the dataset consists earnings of employees of a organization, Mean gives the average salary in the organization.
2. Mode: It gives the most frequent data in a dataset. Suppose the dataset consists the grades of students in a subject, Mode gives you the grade that most student got.
3. Median: It gives the middle value of a dataset when it is sorted properly. Suppose the dataset consists sorted  Marks of students in each subject it gives the mark that is in the middle of the dataset.
