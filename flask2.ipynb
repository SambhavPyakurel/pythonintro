#### Q1. Explain GET and POST methods.

GET method:
GET is used to fetch data from a specified source. It uses urls to do so. So, it's not suitable for confidential data. The response to the GET is cacheable. It is more popular as it is more efficient. Data is transferred in header. So, Small amount of data can only be sent.

POST method:
POST is also used to fetch data from a specified source. It do not uses urls to do so. So, it is more secure. The response to POST is not cacheable. It is less popular. Data is transferred in body. So, large volume of data can be sent.

#### Q2. Why is request used in Flask?

Request is used in flask to retrive request data and to handle the form. It provides a convenient and consistent way to access request data, enabling you to build dynamic and interactive web applications that respond appropriately to client requests.

#### Q3. Why is redirect() used in Flask?

The redirect() function is used to redirect the client's browser to a different URL or route within the application.

#### Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are files that contain HTML (or other markup languages) with placeholders for dynamic content. 

render_template() is used to process the template file, replace the placeholders with the provided data, and generate the final HTML output to be returned to the client.

#### Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask, render_template, jsonify, request

app = Flask(__name__)

@app.route("/Postman_action", methods=['POST'])
def hello_world():
    if request.method == 'POST':
        name = request.json['name']
        roll = request.json['roll']

        return jsonify({'Name': name, 'Roll': roll})

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
