#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and popular web framework written in Python. It provides a simple and flexible way to build web applications and APIs. Flask follows the WSGI (Web Server Gateway Interface) specification and is designed to be easy to understand and use, making it an excellent choice for beginners and small to medium-sized projects.

Advantages

1. Simplicity and minimalism: Flask provides only essential functions needed for web development.
2. Lightweight and flexibility: Flask is framework of small codebase making it lightweight and has modular design that allows developers to add or remove the features.
3. RESTfulness: Flask uses REST architecture.

#### Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the process of defining URL routes in a Flask application. It involves mapping specific URLs to corresponding functions or view handlers that handle the incoming requests and generate responses. The Flask framework provides a decorator called @app.route that allows developers to define routes easily.

App routes are used to:

1. URL mapping: App routing defines the mapping between specific URL to a specific function.
2. Request handling: App routing provides a way to handle the request from every URL by routing it to specific function.
3. Code Organisation: App routing helps develop different sections of the application into distinct function which promotes organization and maintainability the code.

```python
from flask import Flask,url_for

app = Flask(__name__)

@app.route('/welcome/<username>')
def profile(username):
    return f'Welcome, {username}'


if __name__=="__main__":
   # app.run(host="0.0.0.0")
    
    with app.test_request_context():
        user_url = url_for('profile', username='John')
        
        print(f'User URL: {user_url}')
    
```
