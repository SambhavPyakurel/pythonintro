#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Scatter plot is a type of plot that defines the relation between two numbers using a data point.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("This is scatter plot")
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows to arrange multiple plots or axes in a grid-like structure, making it easier to display and compare multiple visualizations or data representations in a single figure.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])  
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, multi = plt.subplots(nrows=2, ncols=2)

multi[0, 0].plot(x1, y1)
multi[0, 0].set_title('Plot1')

multi[0, 1].plot(x2, y2)
multi[0, 1].set_title('Plot2')

multi[1, 0].plot(x3, y3)
multi[1, 0].set_title('Plot3')

multi[1, 1].plot(x4, y4)
multi[1, 1].set_title('Plot4')

plt.tight_layout()
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Bar plot is type of plot used to compare categorical values using bars. It is used to compare categorical values.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel('Companies')
plt.ylabel('Profit')
plt.title('Companies and their profit.')
plt.show()
```

```python
plt.barh(company,profit)
plt.xlabel('Profit')
plt.ylabel('Companies')
plt.title('Companies and their profit.')
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot is a graphical representation of the distribution of a dataset. It summarizes the minimum, first quartile, median, third quartile, and maximum values of a set of numerical data.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
fig,box=plt.subplots()
box.boxplot([box1,box2])
box.set_xticklabels(['Box1','Box2'])
box.set_ylabel('Values')
box.set_title('Box Plot') 
plt.show()
```
