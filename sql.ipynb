#### Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

Data Definition Language is a subset of sql used to specify the content and structure of the table in a database.It is used to define the physical characteristics of records.

```python
--Create: Creates DATABASE

CREATE DATABASE Test;

SHOW DATABASES;
```

```python
--Drop: Deletes a database

DROP DATABASE Test;

SHOW DATABASES;
```

```python
--Alter: Modifies the table.

CREATE DATABASE Test;

CREATE TABLE Test.SubTest(column1 int);

DESC Test.SubTest;

```

```python
ALTER TABLE Test.SubTest
ADD Column2 int;

DESC Test.SubTest;
```

```python
--Truncate: Used to clear all information of a table.

Insert into Test.SubTest(column1,Column2)
VALUES
(1,1),
(2,2);

Select *FROM Test.SubTest;
```

```python
TRUNCATE TABLE Test.SubTest;

SELECT *FROM Test.SubTest;
```

#### Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

Data Manipulation Language is a subset of SQL used for manipultion of records such as Insert,Update,deletion of records etc.

```python
--Insert: Inserts data in a TABLE

INSERT INTO Test.SubTest(column1,Column2)
VALUES
(1,1),
(2,2);

SELECT *from Test.SubTest;
```

```python
--Update: Updates data of a table.

UPDATE Test.SubTest
SET column1=3 WHERE Column2=2;

SELECT *FROM Test.SubTest;
```

```python
--Delete: Deletes record from a table.

DELETE FROM Test.`SubTest`
WHERE column1=1;

SELECT *FROM Test.`SubTest`;
```

#### Q5. Explain Primary Key and Foreign Key.

Primary Key:
A primary key is a column or a set of columns in a database table that uniquely identifies each record or row. It serves as a unique identifier and ensures the uniqueness and integrity of the data within the table. The primary key constraint guarantees that no two rows in the table can have the same values for the primary key column.

Foreign Key:
A foreign key is a column or a set of columns in a database table that refers to the primary key of another table. It establishes a relationship between two tables based on the values of the foreign key column. The foreign key constraint ensures referential integrity and enforces that the values in the foreign key column match the values in the primary key column of the referenced table.

#### Q7. Give the order of execution of SQL clauses in an SQL query.

Order:

1. From : Choose and join tables.
2. Where: Filters base data.
3. Group by: Aggregates base data.
4. Having: filters aggregated data.
5. Select: Returns final data.
6. Order by: Sorts final data.
7. Limit: Limits the returned data to row count.
