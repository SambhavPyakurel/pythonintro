#### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is a base class used for exceptions that occur when a lookup or search operation fails.

```python
#e.g

#KeyError

dict0={'a':'apple', 'b':'banana', 'c':'cat'}

try:
    value=dict0[0]
except KeyError as e:
    print("Key error: ",e)
```

Since, no key '0' exists so this is keyerror.

```python
#IndexError
list0=[0,1,2,3]

try:
    value=list0[7]
except IndexError as e:
    print("Index error:",e)
```

Since, the list is doesnot go to 7th place, so this is a indexerror.

#### Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a built-in exception in Python that is raised when an import statement fails to import a module or a name. It indicates that there was a problem in locating or importing the desired module or name.

ModuleNotFoundError is the exception that pops when the module intended to import doesn't exist.

```python
#e.g.

try:
    import random_module
except ModuleNotFoundError as e:
    print("Module not found:",e)
```
